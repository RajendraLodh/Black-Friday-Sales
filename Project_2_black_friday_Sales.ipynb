```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

1. Merge train and test data set .

```python
df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

2. **Drop user id columns it is not much of use .**

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

3. **Converting a categorical feature gender in 0 and 1 .**

```python
# df['Gender'] = pd.get_dummies(df['Gender'],drop_first=1)
#   or

df['Gender']= df['Gender'].map({'F':0,'M':1})
df.head()
```

4. **handling categorical feature Age column .**

```python
df['Age'].unique()
```

```python
# df['Age] = pd.get_dummies(df['Age'],drop_first=True)

df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,
                           '36-45':4,'46-50':5,
                           '51-55':6,'55+':7})
df.head()
```

5. **handling categorical feature City column .**

```python
df_city = pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
df = pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

5. **handling Missing values .**

* Here we have missing values in category 2 and 3 so we are replacing that mode 
* means which ever category has more occurence we have used that values to replcaing missing values .   

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
# replacing missing values with Mode for catergory 2 

df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2']).mode()[0]
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
# replacing missing values with Mode for catergory 3

df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3']).mode()[0]
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

6. **handling Categorical feature stay_in_current_city_years .**

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
# convert object into integers 

df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

```python
df['B'] = df['B'].astype(int)
```

```python
df['C'] = df['C'].astype(int)
```

```python
df.info()
```
